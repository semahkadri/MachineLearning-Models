**Rgression linaire multiple**

```python
X = car_f.drop('MSRP', axis=1)
y= car_f['MSRP'].values
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=3)
```

### Rgression linaire multiple    avec les valeurs par defauts de `LinearRegression`

```python

linreg2 = LinearRegression() 

linreg2.fit(X_train, y_train) 
y_pred = linreg2.predict(X_test)

print('test_score = ',linreg2.score(X_test,y_test)) 
print('R2 = ',r2_score(y_test, y_pred))
print('MSE = ',mean_squared_error(y_test, y_pred))
print('RMSE = ', np.sqrt(mean_squared_error(y_test, y_pred)))
print('MAE = ', mean_absolute_error(y_test, y_pred))
print('Meadian = ', median_absolute_error(y_test, y_pred))

print('Intercept = ', linreg2.intercept_)
print('Coefficients : ',linreg2.coef_)
```

### Rgression linaire multiple   avec `fit_intercept= False` et `normalize=True`

```python
linreg3 = LinearRegression(fit_intercept= False,normalize=True)
linreg3.fit(X_train, y_train)
y_pred = linreg3.predict(X_test)

print('test_score = ',linreg3.score(X_test,y_test)) 
print('R2 = ',r2_score(y_test, y_pred))
print('MAE = ',mean_squared_error(y_test, y_pred))
print('RMSE = ', np.sqrt(mean_squared_error(y_test, y_pred)))
print('MAE = ', mean_absolute_error(y_test, y_pred))
print('Meadian = ', median_absolute_error(y_test, y_pred))
print(linreg3.intercept_)
print(linreg3.coef_)
```

## Selection de variables avec `SelectKBest`

```python
 kbest = SelectKBest(score_func=f_regression, k=3)
kbest.fit(X_train, y_train)
print("Slection de variables :", kbest.get_support())
print("Scores de variables :", kbest.scores_)
print("Variables slectionnes:", list(X.columns[kbest.get_support()]))
print("Variables supprimes :", list(X.columns[~kbest.get_support()]))
```

```python
X_train = kbest.transform(X_train)
X_test = kbest.transform(X_test)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print('test_score = ',model.score(X_test,y_test)) 
print('R2 = ',r2_score(y_test, y_pred))
print('MAE = ',mean_squared_error(y_test, y_pred))
print('RMSE = ', np.sqrt(mean_squared_error(y_test, y_pred)))
print('MAE = ', mean_absolute_error(y_test, y_pred))
print('Meadian = ', median_absolute_error(y_test, y_pred))
```

## Supprimer les variables fortement corrles

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
print(X_train.shape)
print(X_test.shape)
```

```python
X_train.corr()
```

```python
def correlatedFeatures(dataset, threshold):
    correlated_columns = set()
    correlations = dataset.corr()
    for i in range(len(correlations)):
        for j in range(i):
            if abs(correlations.iloc[i,j]) > threshold:
                correlated_columns.add(correlations.columns[i])
    return correlated_columns
```

```python
cf = correlatedFeatures(X_train, 0.7)
cf
```

```python
X_train = X_train.drop(cf, axis=1)
X_test = X_test.drop(cf, axis=1)
print(X_train.shape)
print(X_test.shape)
```

```python

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print('test_score = ',model.score(X_test,y_test)) 
print('R2 = ',r2_score(y_test, y_pred))
print('MAE = ',mean_squared_error(y_test, y_pred))
print('RMSE = ', np.sqrt(mean_squared_error(y_test, y_pred)))
print('MAE = ', mean_absolute_error(y_test, y_pred))
print('Meadian = ', median_absolute_error(y_test, y_pred))
```

## Slection de variables avec Recursive Feature Elimination (RFE)

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
print(X_train.shape)
print(X_test.shape)
```

```python
model = LinearRegression()
rfe = RFE(model, n_features_to_select=8)
rfe.fit(X_train, y_train)
print("Slection de variables", rfe.support_)
print("Classement de variables", rfe.ranking_)
print("Variables slectionnes :", list(X.columns[rfe.support_]))
```

```python
X_train = rfe.transform(X_train)
X_test = rfe.transform(X_test)
print(X_train.shape)
print(X_test.shape)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print('test_score = ',model.score(X_test,y_test)) 
print('R2 = ',r2_score(y_test, y_pred))
print('MAE = ',mean_squared_error(y_test, y_pred))
print('RMSE = ', np.sqrt(mean_squared_error(y_test, y_pred)))
print('MAE = ', mean_absolute_error(y_test, y_pred))
print('Meadian = ', median_absolute_error(y_test, y_pred))
```
